```python
def separate_data_type(df):
    categorical= []
    continuous = []
    for column in Emp.columns:
        if Emp[column].nunique()<20:
            categorical.append(column)
        else:
            continuous.append(column)
            
    return categorical, continuous
    
categorical,continuous = separate_data_type(Emp)
from tabulate import tabulate
table = [categorical, continuous]
print(tabulate({"Categorical":categorical,
                "continuous": continuous}, headers = ["categorical", "continuous"]))
```

```python
def info_of_cat(col): 
    print(f"Unique values in {col} are: {Emp[col].unique()}") # unique: returns the unique values in the col
    print(f"Mode of {col} is {Emp[col].mode()[0]}")  # mode: returns the mode of the column
    print(f"Number of missing values in {col} is {Emp[col].isnull().sum()}") # isnull().sum() counts the number of null in dataframe
      
```

## Level 1 Analysis

```python
info_of_cat("Attrition")
```

Attrition is the nominal datatype. 

Visual representation 

```python
fig,ax = plt.subplots(1,2,figsize=(10,7))
ax[0].set_title('Count of attrition')
percentage = Emp["Attrition"].value_counts()
labels = list(percentage.index)

sns.countplot(x = Emp["Attrition"], ax = ax[0])
plt.pie(x = percentage, autopct = "%1.1f%%", explode= [0.1,0], labels = labels)
plt.show()
```
